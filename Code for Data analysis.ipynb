EDIT LAST PERFORMANCE COULMN BASED ON MARKS IN SUBJECTS

```python
# now we are converting performance column in two categories 0 and 1 for fet Target value to predict their performance.
def categories(row):
    if row["COA"]>50 and row["OS"]>50 and row["CN"]>50 and row["OOP-1"]>50 and row["DS"]>50 and row["EG"]>50:
        return 1
    else:
        return 0
```

```python
df['Target']=df.apply(lambda row: categories(row),axis=1)
```

```python
df.head()
```

```python
# Now we don't need of performance column in our dataset. so we can drop this column.
df.drop(["PERFORMANCE"],axis=1)
```

# Data visulaization

```python
# here our target column depends on marks in every subject so we are creating charts for them.
sns.pairplot(df, hue='Target')
```

```python
# here we can see that we get extra columns which are not used for target column so first we have to delete them. 
# Sr no. , Enrollment no. , Mobile number
df1 = df.drop(['Enrollment Number','Mobile Number','PERFORMANCE'], axis=1)
```

# DATA SPLIT

```python
df2 = df1.drop(["Name","Gender","City"],axis=1)
X = df2.drop("Target", axis=1)
y = df1["Target"]
print("shape of X=", X.shape)
print("shape of y=", y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=20)
print("shape of X_train=", X_train.shape)
print("shape of y_train=", y_train.shape)
print("shape of X_test=", X_test.shape)
print("shape of y_test=", y_test.shape)
```

# model build

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# COMPARE THREE MODEL AT A SINGLE TIME

```python
X_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay,  \
                            precision_score,recall_score,f1_score,roc_auc_score,roc_curve
```

```python
models = {
    "Logistic Regression":LogisticRegression(),
    "Decision Tree":DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier()
}
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_train)
    
    y_train_pred=model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    model_train_accuracy = accuracy_score(y_train, y_train_pred)
    model_train_f1 = f1_score(y_train, y_train_pred, average='weighted')
    model_train_precision = precision_score(y_train, y_train_pred)
    model_train_recall = recall_score(y_train, y_train_pred)
    model_train_rocauc_score = roc_auc_score(y_train, y_train_pred)
    
    model_test_accuracy = accuracy_score(y_test, y_test_pred)
    model_test_f1 = f1_score(y_test, y_test_pred, average='weighted')
    model_test_precision = precision_score(y_test, y_test_pred)
    model_test_recall = recall_score(y_test, y_test_pred)
    model_test_rocauc_score = roc_auc_score(y_test, y_test_pred)
    
    
    print(list(models.keys())[i])\
    
    print('Model performace for Training set')
    print("- Accuracy : {:.4f}".format(model_train_accuracy))
    print("-F1 score : {:.4f}".format(model_train_f1))
    print("-Precision : {:.4f}".format(model_train_precision))
    print("-Recall : {:.4f}".format(model_train_recall))
    print("-Roc AUc Score: {:.4f}".format(model_train_rocauc_score))
    
    print("-----------------------------------")
    
    print('Model performace for Test set')
    print("- Accuracy : {:.4f}".format(model_test_accuracy))
    print("-F1 score : {:.4f}".format(model_test_f1))
    print("-Precision : {:.4f}".format(model_test_precision))
    print("-Recall : {:.4f}".format(model_test_recall))
    print("-Roc AUc Score: {:.4f}".format(model_test_rocauc_score))
    
    print("="*35)
    print('\n')
```
